```python
path = glob("data/training-test-images/Thermal/2019-02-16/101MEDIA/DJI_0150.jpg")
```

```python
for im in path:
    with rio.open(im) as src:
        image = src.read()
        
        fig,ax = plt.subplots(1,2,figsize=(14,7))
        ax[0].axis('off')
        ax[0].imshow(image[0], cmap = "gray")
        ax[1].hist(image[0].flatten(), bins=100, facecolor='b', edgecolor='k', alpha=0.8)
        ax[1].set(xlabel="Pixel Value", ylabel="Pixel Count")
        plt.show()
```
