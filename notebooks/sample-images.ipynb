```python
import numpy as np
import rasterio as rio
import matplotlib.pyplot as plt
import os
import geopandas as gpd
from glob import glob
```

```python
os.chdir("/Users/micha/ea-applications")
#(os.path.join("Users", "micha", "ea-applications"))
```

```python
bad_sample_path = glob("sample-bad/*.jpg")
#bad_sample_path
```

```python
for image in bad_sample_path:
    with rio.open(image) as src:
        bad = src.read()
        
        fig,ax = plt.subplots(1,2,figsize=(10,5))
        ax[0].imshow(bad[0], cmap = "gray")
        ax[1].hist(bad[0].flatten(), bins=100, facecolor='b', edgecolor='k', alpha=0.15)
        plt.show()
```

```python
good_sample_path = glob("sample-good/*.jpg")
#good_sample_path
```

```python
for image in good_sample_path:
    with rio.open(image) as src:
        good = src.read()
        
        fig,ax = plt.subplots(1,2,figsize=(10,5))
        ax[0].imshow(good[0], cmap = "gray")
        ax[1].hist(good[0].flatten(), bins=100, facecolor='b', edgecolor='k', alpha=0.15)
        plt.show()
```
